```python
# constructor : special method will exists at the time of object creation
# def__init__self()
class student:
    def __init__(self,name,branch):
        print('constructor executed..')
        self.name=name
        self.branch=branch
    def Details(self):
        print('hello:',self.name)
        print('branch is:',self.branch)
        print('method execution..')
```

```python
name=input('enter name:')
branch=input('enter branch:')
s=student(name,branch)
s.Details()
```

```python
### inheritance
-- getting the properties from parent class to child class
-- ha a relation

-- creating the object of parent class to clild class 
--is - a relation
** directly pass the parent class 
```

```python
class parent:
    money=100000
    land=10
class child:
    study='b tech'
    def __init__(self):
        self.prop=parent()
    def properties(self):
        print('money of parent is:',self.prop.money)
        print('land of parent is:',self.prop.land)
        print('child property is:',child.study)
```

```python
c=child()
c.properties()
```

# polymorphasim
-- many forms
-- one will do many operations
-- overloading
      * same name with different parameters
-- overriding
     ** same name and parameters or arguments in different classes 

```python
class poly:
    def m1(self):
        print(' method arg method')
    def m1(self,a):
        print('one arg method')
    def m1(self,a,b):
        print('two arg method')
        print(a+b)
        
        p=poly()
        p.add(6,8)

    
```

```python
print(p)
```

```python
p.add(1,2)
```

```python
class A:
    def m1(self,a,b):
        print(a*b)
class B:
    def m1(self,a,b):
        print(a+b)
a=A()
b=B()

a.m1(2,5)
b.m1(4,5)
```
